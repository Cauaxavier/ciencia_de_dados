### Adicionar 1 coluna

```python
# a partir de uma coluna que existe
tabela['Comissão'] = tabela['Valor Final'] * 0.05
#display(tabela)

# criar uma coluna com valor padrão
# não faça assim: tabela['Imposto']=0
tabela.loc[:, 'Imposto'] = 0
display(tabela)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
vendas_dez = pd.read_excel("Vendas - Dez.xlsx")
vendas_dez['Comissão'] = vendas_dez['Valor Final'] * 0.05
vendas_dez.loc[:, 'Imposto'] = 0
display(vendas_dez)
#tabela = tabela.append(vendas_dez) o baguio ta obsolet e eu não sei substituir
#display(tabela)
```

### Excluir linhas e colunas

```python
tabela = tabela.drop('Imposto', axis=1)
display(tabela)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# Deletar linhas ou colunas vazias
#tabela = tabela.dropna(how=all, axis=1)

# Deletar linhas que possuem valores vazios
#tabela = tabela.dropna()

# PREENCHER COM VALORES VAZIOS

# preencher com a média da coluna
# tabela['Comissão'] = tabela['Comissão'].fillna(tabela['Comissão'].mean())
# tabela.loc[tabela['Comissão'] == 0, 'Comissão'] = tabela['Comissão'].mean()
# tabela = tabela.round(2)

# preencher com o ultimo valor

tabela = tabela.ffill()
display(tabela)
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
tabela = tabela.drop_duplicates()
display(tabela)

# value counts
transacoes_loja = tabela['ID Loja'].value_counts()
display(transacoes_loja)

#agrupando colunas
faturamento_Produto = tabela[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_Produto)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
gerentes = pd.read_excel('Gerentes.xlsx')
display(gerentes)
tabela = tabela.merge(gerentes)
display(tabela)
```
